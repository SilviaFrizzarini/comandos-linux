Instalar bcftools com plugin (+split-vep)

```python
#Código na célula será interpretado como comandos Bash.
%%bash
#Clona o repositório htslib do GitHub + submódulos.
git clone --recurse-submodules https://github.com/samtools/htslib.git
#Clona o repositório bcftools do GitHub.
git clone https://github.com/samtools/bcftools.git
#Altera o diretório de trabalho para o diretório bcftools.
cd bcftools
#Compila o código-fonte do bcftools.
make
#Instala o programa compilado no sistema.
make install
```

# **WP017**

```python
#WP017
#Código na célula será interpretado como comandos Bash.
%%bash
#Extrai as quatro primeiras colunas do arquivo, substitui o texto CHROM por CHR nas colunas extraídas, salva o resultado no arquivo e exisbe as 10 primeiras linhas do arquivo.
cut -f1-4 /content/lmabrasil-hg38/vep_output/liftOver_WP017_hg19ToHg38.vep.filter.tsv | sed -e "s/CHROM/CHR/g"  > df_WP017-cgi.txt
head df_WP017-cgi.txt
```

Enviando um Job - job_id.

```python
import requests
#Contém informações de autenticação.
headers = {'Authorization': 'silfrizzarini@gmail.com 7f1b6ddf77bc5186c964'}
#Dicionário que contém os dados a serem enviados no corpo da requisição.
payload = {'cancer_type': 'HEMATO', 'title': 'Somatic MF', 'reference': 'hg38'}
#Envia a requisição.
r = requests.post('https://www.cancergenomeinterpreter.org/api/v1',
                headers=headers,
                files={
                        'mutations': open('/content/df_WP017-cgi.txt', 'rb')
                        },
                data=payload)
#A resposta da requisição é armazenada na variável r. Converte a resposta em formato JSON.
r.json()
```

Acessando informações do Job.

Download dos Resultados (file.zip).

```python
#Baixa os resultados da análise em file.zip, fica disponível no diretório de trabalho.
import requests
job_id ="0600109548f9620e87e4"

headers = {'Authorization': 'silfrizzarini@gmail.com 7f1b6ddf77bc5186c964'}
payload={'action':'download'}
r = requests.get('https://www.cancergenomeinterpreter.org/api/v1/%s' % job_id, headers=headers, params=payload)
with open('file.zip', 'wb') as fd:
    fd.write(r._content)
```
